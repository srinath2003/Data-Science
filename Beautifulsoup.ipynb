```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP request to the webpage
url = 'https://www.simplilearn.com/data-wrangling-article#:~:text=Data%20wrangling%20is%20the%20process%20of%20removing%20errors%20and%20combining,analysis%20is%20becoming%20increasingly%20necessary.'
response = requests.get(url)
var=response.text
var

```

```python
f= open('filename_webscrape.txt', 'w',encoding='utf-8')
f.write(var)
```

```python
# Create a Beautiful Soup object to parse the HTML
soup = BeautifulSoup(response.text, 'html.parser')
type(soup)

```

```python
# Extract specific data
title = soup.title.text
paragraphs = soup.find_all('p')
type(paragraphs)
```

```python
# Print the extracted data
print("Title:", title)
for p in paragraphs:
    print("Paragraph:", p.text)
```
