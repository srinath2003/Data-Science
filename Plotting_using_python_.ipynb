```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
op=pd.read_csv("/content/accidental-deaths-in-usa-monthly (1).csv")


#op.info
op.describe


op.duplicated().sum()
sb.boxplot(data=op)
sb.violinplot(data=op)
op.plot(kind = 'scatter',x="Month",y="Accidental deaths in USA: monthly, 1973 ? 1978")
op.plot(kind='hist')

op.plot()

op.plot.bar()


op.plot.barh(stacked=True)

plt.hist(x="Accidental deaths in USA: monthly, 1973 ? 1978")
plt.show()

op.plot.area()



```
